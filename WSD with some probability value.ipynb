```python
def computeTF(wordDict,bow):
    tfDict={}
    bowCount=len(bow)
    for word,count in wordDict.items():
        #TF=No. of words in a sentence / total words in a sentence
        tfDict[word]=count/float(bowCount)
    return tfDict
```

```python
tfBowA=computeTF(wordDictA,X_11)
tfBowB=computeTF(wordDictB,X_22)
tfBowC=computeTF(wordDictC,X_33)
tfBowD=computeTF(wordDictC,X_44)
```

```python
#creating the log portion of the Excel table we saw earlier
import math
def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10((N+1) / float(val)+1.0)+1
        
    return idfDict
#inputing our sentences in the log file
idfs = computeIDF([wordDictA, wordDictB,wordDictC,wordDictD])
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
#running our two sentences through the IDF:
idfFirst = computeTFIDF(tfBowA, idfs)
idfSecond = computeTFIDF(tfBowB, idfs)
idfThird = computeTFIDF(tfBowC, idfs)
idfFourth=computeTFIDF(tfBowD,idfs)
#putting it in a dataframe
idf= pd.DataFrame([idfFirst, idfSecond,idfThird,idfFourth])
#idf
```
